```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Reading the dataset
df=pd.read_csv("amazon.csv")
```

```python
print("First 5 rows of the dataset")
df.head()
```

```python
#Checking for missing values and Getting the information
print("Information about the dataset")
df.info()
print("\nChecking for missing values")
df.isnull().sum()
```

```python
print("\nRows with null values:\n", df[df.isnull().any(axis=1)])
```

```python
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')  # Convert to numeric
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].mean())  # Assigning back
```

```python
print("\nNull Values After Cleaning:\n", df.isnull().sum())
```

```python
# Create a new DataFrame for sales analysis
sales_df = df[['product_name', 'category', 'discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].copy()

# Convert price columns to numeric (remove ₹ and commas)
sales_df['discounted_price'] = sales_df['discounted_price'].replace('[₹,]', '', regex=True).astype(float)
sales_df['actual_price'] = sales_df['actual_price'].replace('[₹,]', '', regex=True).astype(float)
sales_df['discount_percentage'] = sales_df['discount_percentage'].replace('%', '', regex=True).astype(float)

# Convert ratings and rating count to numeric
sales_df['rating'] = pd.to_numeric(sales_df['rating'], errors='coerce')
sales_df['rating_count'] = pd.to_numeric(sales_df['rating_count'], errors='coerce').fillna(0)

# Calculate total sales revenue (estimated)
sales_df['estimated_revenue'] = sales_df['discounted_price'] * sales_df['rating_count']

# Display summary statistics
print("Sales Data Overview:\n", sales_df.describe())
# Save the cleaned sales data
sales_df.to_csv("cleaned_sales_data.csv", index=False)
print("\nCleaned Sales Data Saved as 'cleaned_sales_data.csv'")
```

```python
#Data Visualization
#Distribution of Discount Percentages
plt.figure(figsize=(8,5))
sns.histplot(sales_df['discount_percentage'], bins=20, kde=True, color='blue')
plt.title("Distribution of Discount Percentages")
plt.xlabel("Discount Percentage")
plt.ylabel("Count")
plt.show()
```

```python
#Relationship between Discount and Sales (Rating Count)
plt.figure(figsize=(8,5))
sns.scatterplot(x='discount_percentage', y='rating_count', data=sales_df, alpha=0.7)
plt.title("Effect of Discount on Sales")
plt.xlabel("Discount Percentage")
plt.ylabel("Number of Ratings (Sales Proxy)")
plt.show()
```

```python
#Relationship Between Discount and Price
plt.figure(figsize=(8,5))
sns.lineplot(x=sales_df['discount_percentage'], y=sales_df['discounted_price'], color='red')
plt.title("Relationship Between Discount Percentage and Discounted Price", fontsize=14)
plt.xlabel("Discount Percentage", fontsize=12)
plt.ylabel("Discounted Price", fontsize=12)
plt.show()
```

```python
#Distribution of Discounted Prices
plt.figure(figsize=(8,5))
sns.histplot(sales_df['discounted_price'], bins=30, kde=True, color='blue')
plt.title("Distribution of Discounted Prices", fontsize=14)
plt.xlabel("Discounted Price", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
# Market Share of Top 5 Categories (Based on Count)
top_category_counts = sales_df['category'].value_counts().nlargest(5)
plt.figure(figsize=(7,7))
plt.pie(top_category_counts, labels=top_category_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Market Share of Top 5 Categories", fontsize=14)
plt.show()
```

```python
# Create Bubble Plot
plt.figure(figsize=(10, 6))
plt.scatter(
    sales_df['actual_price'],
    sales_df['discounted_price'],
    edgecolors='w'
)

plt.xlabel("Actual Price (₹)", fontsize=12)
plt.ylabel("Discounted Price (₹)", fontsize=12)
plt.title("Bubble Plot: Actual Price vs Discounted Price ", fontsize=14)
plt.grid(True)
plt.show()
```
